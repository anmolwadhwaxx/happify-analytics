```python
import seaborn
import pandas
import matplotlib.pyplot as plt
csv = pandas.read_csv(r"C:\Users\Vaibhavi\responses - Copy.csv")
res = seaborn.boxenplot(x=csv['Age'])
plt.show()
```

INFERENCE : A major pool of students lie in the age group of 19 to 21 with 18, 22, 23, 24 and 25 being outliers.

```python
res = seaborn.countplot(x=csv['Age'])
plt.show()
```

```python
res = seaborn.countplot(x=csv['Gender'])
plt.show()
```

INFERENCE : Population of females who took part in the survey was less than half of male population.

```python
mapping1 = {1:'Highly Dissappointed(1)', 2:'Dissappointed(2)', 3:'Satisfied(3)', 4:'Higly Satisfied(4)'}
csv = csv.replace({'On a scale of 1 to 4, how much are you satisfied with your academic performance?': mapping1})

mapping2 = {0:'Stable', 1:'Stressed', 2:'Critical'}
csv = csv.replace({'Category in which students identified themselves': mapping2})
```

```python
seaborn.set_style('whitegrid')
ax=seaborn.countplot(x='On a scale of 1 to 4, how much are you satisfied with your academic performance?', hue ='Gender', data=csv, order=['Highly Dissappointed(1)','Dissappointed(2)', 'Satisfied(3)', 'Higly Satisfied(4)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
seaborn.set_style('whitegrid')
ax=seaborn.countplot(x='On a scale of 1 to 4, how much are you satisfied with your academic performance?', data=csv, order=['Highly Dissappointed(1)','Dissappointed(2)', 'Satisfied(3)', 'Higly Satisfied(4)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

INFERENCE : It was found in the survey that maximum students were moderately satisfied with their academic performance.

```python
csv['Category in which students identified themselves'].value_counts()
```

```python
data0 = list(csv['Category in which students identified themselves'].value_counts())
keys0 = ['Stressed', 'Stable', 'Critical']

explode = [0, 0, 0.1]
palette_color = seaborn.color_palette('dark')
plt.pie(data0, labels=keys0, colors=palette_color,
        explode=explode, autopct='%.0f%%')
plt.title('Category waise distribution of students')
plt.show()
```

INFERENCE : More than 50% of the students who attempted the survey identified themselves as stressed.

```python
seaborn.set_style('whitegrid')
seaborn.countplot(x='Category in which students identified themselves',hue='Gender',data=csv,palette='RdBu_r', order=['Stable', 'Stressed', 'Critical'])
```

INFERENCE : Even amongst the small population of female participants, those who had critical condition were almost equal in number as compared to male counterparts in the same category.

```python
res = seaborn.catplot(x="Time spent on digital entertainment", y="Category in which students identified themselves", data=csv)
plt.show()
```
